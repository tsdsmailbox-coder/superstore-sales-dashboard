```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\borin\Downloads\superstore.csv")    #load data

df.head()         #shows first 5 rows
df.info()         #summary info 
df.isnull().sum()   #counts missing/emptyb
```

```python
#which category makes the most sales
df.groupby ("Category")["Sales"].sum().sort_values(ascending=False).head(5)        #groups data by category and calculate total sales for each category and sorts from hightest to lowest and shows top 5 categories     
```

```python
#which year has the highest revenue
df["Order.Date"] = pd.to_datetime(df["Order.Date"])         #convert order date into df so panda can work with dates
df["Year"] = df["Order.Date"].dt.year                       #extracts year number from order date ans store in new column year
df.groupby("Year")["Sales"].sum()                           #groups data by year and calculates the total sales for each year
```
